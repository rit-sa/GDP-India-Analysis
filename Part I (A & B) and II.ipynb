```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
import glob
```

```python
all_files = glob.glob("D:\\upgrad\\sgdp\\*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename,encoding = "ISO-8859-1")
    df = df.loc[::,['Item','2014-15']]
    df.columns = df.columns.str.replace('*','').str.replace('*','')
    df = df.set_index(['Item']).T
    df['State'] = filename.split('-')[1]
    df= df.set_index('State')
    li.append(df)
pp = pd.concat(li, axis=0)
pp.head()
```

```python
ssk =pp.T
ssk.head(48)
```

```python
rd = ssk.iloc[[9,15,17,24,31], :]
rd.head()
```

```python
#Gross State Domestic Product for 2014-15 
# Per Capita GSDP Rs. for 2014-15
oo =rd.T
oo.head(2)
```

```python
#Top 5 state as per Per Capita GSDP (Rs.)
vb= oo.sort_values("Per Capita GSDP (Rs.)",ascending= False).head(5)
vb.head()
```

```python
plt.figure(figsize= (10,10))
sns.set_style("whitegrid")
sns.barplot(y=vb["Per Capita GSDP (Rs.)"], x=vb.index, data=vb)
plt.ylabel("Per Capita GSDP (Rs.)")
plt.title("Top 5 States as GDP per capita")
plt.xlabel("State")
plt.show()
```

```python
#Bottom 5 state as per Per Capita GSDP (Rs.)
vv= oo.sort_values("Per Capita GSDP (Rs.)",ascending= True).head(5)
vv.head(38)
```

```python
plt.figure(figsize= (10,10))
sns.set_style("whitegrid")
sns.barplot(y=vv["Per Capita GSDP (Rs.)"], x=vv.index, data=vv)
plt.ylabel("Per Capita GSDP (Rs.)")
plt.title("Bottom 5 States as GDP per capita")
plt.xlabel("State")
plt.show()
```

```python
# Ratio of highest per capita GDP to the lowest per capita GDP.
aa = max(oo["Per Capita GSDP (Rs.)"])/min(oo["Per Capita GSDP (Rs.)"])
print(aa)
```

```python
# Create a new variable
pf =rd.T
pf.head(2)
```

```python
# Create Primary percentage

pf["Primary %"] = pf['Primary']/pf['Gross State Domestic Product']*100
pf.head(2)
```

```python
# Create Secondary percentage

pf["Secondary %"] = pf['Secondary']/pf['Gross State Domestic Product']*100
pf.head(2)
```

```python
# Create Tertiary percentage
pf["Tertiary %"] = pf['Tertiary']/pf['Gross State Domestic Product']*100
pf.head()
```

```python
pf1= pf.drop(columns=["Gross State Domestic Product", "Per Capita GSDP (Rs.)","Primary", "Secondary", "Tertiary"])
pf1.head()
```

```python
# Plotting the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.
plt.figure(figsize=(11,11))
plt.title("Percentage Contribution of Primary, Secondary and Tertiary", size = 19)
pf2 = sns.heatmap(pf1, annot = True)
```

```python
cc =rd.T
cc.head(2)
```

```python
# GDP per capita (C1, C2, C3, C4)
cc["Percentile"] = cc["Per Capita GSDP (Rs.)"].rank(pct = True)
cc.head()
```

```python
#It has help to create state wise catgory
cc["Category"] = pd.qcut(cc["Per Capita GSDP (Rs.)"],[0,0.20,0.50,0.85,1], labels= ["C4","C3","C2","C1"])
cc.head()
```

```python
df3 = pd.merge(pp,cc, on = "State")
df3.head(3)
```

```python
df3.sort_values(by="Category",ascending = False).head()
```

```python
df4 = df3[["Category", "Agriculture, forestry and fishing", "Mining and quarrying","Manufacturing","Electricity, gas, water supply & other utility services", "Construction", "Trade, repair, hotels and restaurants", "Transport, storage, communication & services related to broadcasting", "Financial services", "Real estate, ownership of dwelling & professional services", "Public administration", "Other services","Gross State Domestic Product_x"]]
df4
```

```python
# Shorting by Category and Gross State Domestic Product
ty = df4.sort_values(by = ["Category","Gross State Domestic Product_x"], ascending = False)
ty.head()
```

```python
df5 = df4.groupby("Category").sum()
df5.head()
```

```python
#df5 = df4.set_index("Category")
#df5.head(70)

df6 =df5.T
df6.head(14)
```

```python
#  The sub-sectors contribution as a percentage of the GSDP of each category.

plt.figure(figsize=(13,12), dpi =90)
plt.title("The sub-sectors contribution as a percentage of the GSDP.", size = 17)
df7 = sns.heatmap(df6, annot = True, fmt=".1f", linewidths= 1, square = True, cmap="Blues_r")
```
